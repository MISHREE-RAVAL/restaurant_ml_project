```python
# Handling missing values
data.fillna({'Average Cost for two': 0, 'Cuisines': 'Unknown'}, inplace=True)
# Checking for missing values
data.isnull().sum()
```

```python
#removing duplicate entries of same restaurant name
dfR = dfR.drop_duplicates('Restaurant Name',keep='first')
dfR
```

```python
dfR = dfR[dfR['Aggregate rating']>3.9]
dfR
```

```python
# Encoding categorical variables
label_encoder = LabelEncoder()
data['Cuisines'] = label_encoder.fit_transform(data['Cuisines'])
# Checking the encoding
data[['Cuisines']].head()
```

```python
# Defining criteria for restaurant recommendation
criteria = ['Cuisines', 'Average Cost for two', 'Price range']
user_preferences = {
    'Cuisines': 'Italian',  # Example preference
    'Average Cost for two': 500,  # Example preference
    'Price range': 2  # Example preference
}
```

The criteria selected are Cuisines, Average Cost for two, and Price range, which are commonly used in restaurant recommendations.
user_preferences dictionary is an example of the user's preferences.

```python
# Converting Cuisines into numeric form using TfidfVectorizer
vectorizer = TfidfVectorizer(stop_words='english')
cuisines_matrix = vectorizer.fit_transform(data['Cuisines'].astype(str))
```

TfidfVectorizer is used to convert Cuisines into a sparse matrix of term frequencies.
This helps in comparing the similarity between different restaurants based on their cuisines.



```python
# Calculating cosine similarity between restaurants based on Cuisines
cosine_sim = cosine_similarity(cuisines_matrix, cuisines_matrix)
```

```python
# Recommendation function based on user preferences
def recommend_restaurants(user_preferences, data, cosine_sim):
    user_cuisine_idx = data[data['Cuisines'] == label_encoder.transform([user_preferences['Cuisines']])[0]].index[0]
    cosine_scores = cosine_sim[user_cuisine_idx]

    # Sorting restaurants by similarity
    similar_restaurants = list(enumerate(cosine_scores))
    similar_restaurants = sorted(similar_restaurants, key=lambda x: x[1], reverse=True)

    recommendations = []
    for i in range(5):  # Top 5 recommendations
        restaurant_idx = similar_restaurants[i][0]
        recommendations.append(data['Restaurant Name'][restaurant_idx])

    return recommendations

# Running the recommendation function
recommended_restaurants = recommend_restaurants(user_preferences, data, cosine_sim)
recommended_restaurants
```

```python
# Sample preferences
sample_preferences = {
    'Cuisines': 'Chinese',
    'Average Cost for two': 300,
    'Price range': 1
}

# Running the function with sample preferences
sample_recommendations = recommend_restaurants(sample_preferences, data, cosine_sim)
sample_recommendations
```

```python
# Sample user preferences (Example 1)
sample_preferences_1 = {
    'Cuisines': 'Italian',  # User prefers Italian cuisine
    'Average Cost for two': 500,  # User prefers restaurants with an average cost around 500
    'Price range': 2  # User prefers restaurants with medium price range (Price range: 2)
}

# Running the recommendation function for Example 1
recommended_restaurants_1 = recommend_restaurants(sample_preferences_1, data, cosine_sim)

# Outputting the top 5 restaurant recommendations for Example 1
print("Top 5 Recommendations for User 1 (Italian Cuisine):")
for idx, restaurant in enumerate(recommended_restaurants_1, start=1):
    print(f"{idx}. {restaurant}")


# Sample user preferences (Example 2)
sample_preferences_2 = {
    'Cuisines': 'Chinese',  # User prefers Chinese cuisine
    'Average Cost for two': 300,  # User prefers restaurants with an average cost around 300
    'Price range': 1  # User prefers restaurants with lower price range (Price range: 1)
}

# Running the recommendation function for Example 2
recommended_restaurants_2 = recommend_restaurants(sample_preferences_2, data, cosine_sim)

# Outputting the top 5 restaurant recommendations for Example 2
print("\nTop 5 Recommendations for User 2 (Chinese Cuisine):")
for idx, restaurant in enumerate(recommended_restaurants_2, start=1):
    print(f"{idx}. {restaurant}")
```

```python
from sklearn.metrics import jaccard_score

restaurant_name = 'Ooma'
restaurant_data = data[['Cuisines', 'Average Cost for two', 'Price range']].values.tolist()
target_restaurant_index = data[data['Restaurant Name'] == restaurant_name].index[0]

jaccard_similarities = []
for i, restaurant_features in enumerate(restaurant_data):
    similarity = jaccard_score(restaurant_data[target_restaurant_index], restaurant_features, average='weighted')
    jaccard_similarities.append(similarity)

dfJaccard = pd.DataFrame({'Restaurant Name': data['Restaurant Name'], 'simScore': jaccard_similarities})
dfJaccard = dfJaccard.set_index('Restaurant Name')

similarities = dfJaccard.loc[restaurant_name].sort_values(ascending=False)
similarities_df = pd.DataFrame({'Restaurant Name': similarities.index, 'simScore': similarities.values})

recommended_restaurants = similarities_df[(similarities_df['Restaurant Name'] != restaurant_name) &
                                          (similarities_df['simScore'] >= 0.7)].head(5)

recommended_with_ratings = pd.merge(recommended_restaurants, dfR[['Restaurant Name', 'Aggregate rating']],
                                    how='inner', on='Restaurant Name')

final_recommendations = recommended_with_ratings.sort_values('Aggregate rating', ascending=False).drop_duplicates('Restaurant Name', keep='first')

final_recommendations
```
